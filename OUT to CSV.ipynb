```python
import csv
import glob
import os

txt_files = glob.glob('./Cluster_Output/4.5e9_1000_Rcrit6_NoOverlapChem_Tides/*.out')

for txt_file in txt_files:
    with open(txt_file, "rt") as input_file:
        in_txt = csv.reader(input_file, delimiter=",")
        #Skip first line of file
        next(in_txt)
        #next(in_txt)
        #print(in_txt)
        filename = os.path.splitext(os.path.basename(txt_file))[0] + '.csv'
        
        print(filename)
        with open(os.path.join('./Cluster_Output/4.5e9_1000_Rcrit6_NoOverlapChem_Tides/', filename), 'wt') as output_file:
            out_csv = csv.writer(output_file)
            out_csv.writerows(in_txt)
```
